#### 2. Use Python time library to estimate the runtime for inputs of $n = [1, 10, 10^2, 10^3, 10^4, 10^5, 10^6, 10^7, 10^8, 10^9]$

````python
from time import time

def timer_func(func):
    def wrap_func(*args, **kwargs):
        t1 = time()*10**12
        result = func(*args, **kwargs)
        t2 = time()*10**12
        return result, t2-t1
    return wrap_func


@timer_func
def sum1(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

@timer_func
def sum2(n):
    total = n*(n+1)//2
    return total

````

````python
runtime1 = [sum1(x)[1] for x in [1, 10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**9]]
print(runtime1)   #Devuelve el tiempo de ejecuci'on para la funcion sum1
````

````python
runtime2 = [sum2(x)[1] for x in [1, 10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**9]]
print(runtime2)  #Devuelve el tiempo de ejecuci'on para la funcion sum2
````
